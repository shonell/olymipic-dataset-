```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wine=pd.read_csv('winequality-red.csv')
wine.head()
```

```python
#Check the distributions
wine.hist(figsize=(12,10));
```

```python
#Dive in to the alcohol content
wine['alcohol'].hist();
```

```python
#replace alcohol values with over 10 alcohol with 1 and any value below with 0
wine['alcohol_class'] = np.where(wine['alcohol']>=10.0, '1', '0')
```

```python
ind=wine['alcohol_class'].value_counts().index.sort_values()
wine['alcohol_class'].value_counts()[ind].plot(kind='bar', rot=0);
```

```python
#Segment the fixed acidity column into three groups. The  label of 0 will be assigned to values (47), 1 will be assigned to 
# values (79), and 2 will be assigned to values (916)
wine.loc[(wine['fixed acidity']>4.0) & (wine['fixed acidity']<=7.0), 'acidity_class'] = 0
wine.loc[(wine['fixed acidity']>7.0) & (wine['fixed acidity']<=9.0), 'acidity_class'] = 1
wine.loc[(wine['fixed acidity']>9.0) & (wine['fixed acidity']<=16.0), 'acidity_class'] = 2
```

```python
ind2=wine['acidity_class'].value_counts().index.sort_values()
wine.acidity_class.map(int).value_counts()[ind2].plot(kind='bar', rot=0);
```
